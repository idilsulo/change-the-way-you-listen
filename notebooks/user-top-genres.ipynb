```python
import spotipy
from spotipy.oauth2 import SpotifyOAuth
import pandas as pd
```

# User Authentication

```python
scope = 'user-top-read'
sp = spotipy.Spotify(client_credentials_manager=SpotifyOAuth(scope=scope))
```

# User Top Genres (Based on Top Artists)
Time_range -> Over what time frame the affinities are computed. Valid values: long_term (calculated from several years of data and including all new data as it becomes available), medium_term (approximately last 6 months), short_term (approximately last 4 weeks).

## Medium Term

```python
results = sp.current_user_top_artists(time_range='medium_term', limit=50)
```

```python
all_genres = [genre for r in results['items'] for genre in r['genres'] ]
```

```python
all_genres
```

```python
from collections import Counter
top_genres = Counter(all_genres)
top_genres = {key : value for key, value in sorted(top_genres.items(), key=lambda k: k[1], reverse=True)}
top_genres
```

```python
top_genres = {key : [value] for key, value in sorted(top_genres.items(), key=lambda k: k[1], reverse=True)}
top_genres_df = pd.DataFrame(top_genres)
top_genres_df
```

```python
top_genres_df = top_genres_df.transpose()
```

```python
top_genres_df.rename(columns = {0: 'count'}, inplace=True)
top_genres_df
```

```python
top_genres_df.to_csv('medium_term_genres.csv')
```

```python
def lookup(genre, genre_lookup):
    if genre in genre_lookup.keys():
        return genre_lookup[genre]
    else:
        return 'other'
```

```python
import operator
def cluster_genres(micro_genres, genre_lookup):
    cluster = {}
    for micro_genre in micro_genres:
        genre = lookup(micro_genre, genre_lookup)
        if genre in cluster.keys():
            cluster[genre] += 1
        else:
            cluster[genre] = 1
    
    return max(cluster.items(), key=operator.itemgetter(1))[0]
        
```

```python
def insert_items(obj, genre, micro_genre, artist):
    genre_item = None
    for child in obj['children']:
        if child['name'] == genre:
            genre_item = child
            break
            
    if not genre_item:
        genre_item = {"name": genre, "children": []}
        obj['children'].append(genre_item)
    
    micro_genre_item = None
    for child in genre_item['children']:
        if child['name'] == micro_genre:
            micro_genre_item = child
            break
    
    if not micro_genre_item:
        micro_genre_item = {"name": micro_genre, "children": []}
        genre_item['children'].append(micro_genre_item)
        
    artist_item = None
    for child in micro_genre_item['children']:
        if child['name'] == artist:
            artist_item = child
            break
    
    if not artist_item:
        artist_item = {"name": artist, "size": 1}
        micro_genre_item['children'].append(artist_item)
    else:
        artist_item['size'] += 1
    
```
