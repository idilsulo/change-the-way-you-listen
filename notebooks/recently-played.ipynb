```python
import spotipy
from spotipy.oauth2 import SpotifyOAuth
```

```python
import pandas as pd
```

```python
scope = 'user-read-recently-played'
sp = spotipy.Spotify(client_credentials_manager=SpotifyOAuth(scope=scope))
```

```python
recently_played = [{'id' : t['track']['id'], 
                    'name' : t['track']['name'], 
                    'played_at' : t['played_at'],
                     'artist_name' : t['track']['artists'][0]['name'],
                     'artist_id' : t['track']['artists'][0]['id']} 
                   for t in sp.current_user_recently_played()['items'] if 'track' in t.keys()]

recently_played_ids = [t['track']['id'] for t in sp.current_user_recently_played()['items'] if 'track' in t.keys()]

df = pd.DataFrame(recently_played)
df
```

```python
features = sp.audio_features(recently_played_ids)
features
```

```python
df_features = pd.DataFrame(features)
df_features.columns.values
```

```python
df_features = df_features.filter(['danceability', 'energy', 'loudness', 'speechiness', 
                                'acousticness', 'instrumentalness', 'liveness',
                                'valence', 'tempo', 'id', 'name'])
df_features
```

```python
sp.current_user_recently_played()['items'][0].keys()
```

```python
df = pd.merge(df, df_features, on='id')
df
```

```python
recent_artists = df['artist_id'].drop_duplicates().tolist()
recent_artists
```

```python
df.plot(x='played_at', y='loudness')
```
